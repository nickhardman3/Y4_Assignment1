```python
import pandas as pd
import matplotlib.pyplot as plt

fig2 = pd.read_csv("Figure2.csv")

fig, ax1 = plt.subplots(figsize=(6,4))
ax1.set_xlabel("Monte Carlo Steps (MCS)")
ax1.set_ylabel("Reduced Energy (U/)", color="red")
ax1.plot(fig2["MCS"], fig2["Energy"], color="red", label="Energy (U/)")
ax1.tick_params(axis="y", labelcolor="red")

ax2 = ax1.twinx()
ax2.set_ylabel("Order Parameter S", color="blue")
ax2.plot(fig2["MCS"], fig2["Order"], color="blue", label="Order Parameter S")
ax2.tick_params(axis="y", labelcolor="blue")

plt.title("Figure 2: U/ and S vs MCS (T* = 0.65)")
fig.tight_layout()
plt.show()

# ========================== Figure 3 ==========================
fig3 = pd.read_csv("Figure3.csv").dropna()
fig3 = fig3.sort_values("Temperature")

plt.figure(figsize=(6,4))
plt.errorbar(
    fig3["Temperature"],
    fig3["OrderMean"],
    yerr=fig3["OrderStd"],
    fmt="o-",
    color="green",
    ecolor="green",
    elinewidth=1,
    capsize=3,
    markersize=5,
)
plt.xlabel("Reduced Temperature, T*")
plt.ylabel("Order Parameter, S")
plt.title("20  20 LebwohlLasher model")
plt.xlim(0.25, 1.6)
plt.ylim(0.3, 1.0)
plt.tight_layout()
plt.show()

```
